```python
from itertools import groupby
from inspect import formatargvalues
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize']=(10,10)

df1=pd.read_csv("/content/drive/MyDrive/Data/Bengaluru_House_Data.csv")
df1.shape
df1.head()
df1.groupby("area_type")["area_type"].agg('count')
df2=df1.drop(["area_type","society","balcony","availability"],axis='columns')
df2.head()
df2.isnull().sum()
df3=df2.dropna()
df3.isnull().sum()
df3['size'].unique()
df3['BHK']=df3['size'].apply(lambda x:int(x.split(' ')[0]))
df3['BHK'].unique()
df3.total_sqft.unique()

def is_float(x):

  try:
    float(x)
  except:
       return False
  return True
df3[~df3['total_sqft'].apply(is_float)]

def gives_a_num(x):
  token=x.split("-")
  if len(token)==2:
    return (float(token[0])+float(token[1]))/2
  try:                                        # this try statement indicates or we can say that if the value of x is already a number then just covernt it into float .
    return float(x)
  except:
    return None
df4=df3.copy()
df4['total_sqft']=df4['total_sqft'].apply(gives_a_num)
# print(df4.loc[30])

df5=df4.copy()
df5["price_per_sqft"]=df5["price"]*100000/df5['total_sqft']
df5.head(10)
len(df5.location.unique())  # here we have applied the len function to get the count

df5.location=df5.location.apply(lambda x:x.strip())

location_status=df5.groupby('location')['location'].agg('count').sort_values(ascending=False) # understand it once again
# print(location_status)
len(location_status[location_status<=10])

location_less_than_10=location_status[location_status<=10]
# print(location_less_than_10)

len(df5.location.unique())
df5.location=df5.location.apply(lambda x: "other" if x in location_less_than_10 else x)
len(df5.location.unique())
df5.head(10) # till here

# Outlier And Removal :-  Outlier And Removal can be removed with the following technique as follow :-
# 1.Standard Deviation
# 2. Simple Domain Knowledge (here Domain means "Domain of Project like for eg:- House Prediciton is the project and Real Estate becomes the Domain.")


df5[df5.total_sqft/df5.BHK<300].head() # so will show the output in which the number of BHK comparetively more then the number of total_sqft
# print(df5.shape) #(13246, 7)
df6=df5[~(df5.total_sqft/df5.BHK<300)] # here we have filter the wrong the data using "~" and storing it into a new dataframe
df6.shape #(12502, 7)
df6.price_per_sqft.describe()

# just a look video on standard deviation.

def remove_pps_outlier(df):
  df_out=pd.DataFrame()
  for key, subdf in df.groupby('location'):
    m=np.mean(subdf.price_per_sqft)
    st=np.std(subdf.price_per_sqft)
    reduced_df=subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
    df_out=pd.concat([df_out,reduced_df],ignore_index=True)
  return df_out

df7=remove_pps_outlier(df6)
df7.shape#(10241, 7)

# -------------------------------------------------------------------------------------------------------------
# def scatter_plot(df,location):
#   bhk2=df[(df.location==location) & (df.BHK==2)]
#   bhk3=df[(df.location==location) & (df.BHK==3)]
#   matplotlib.rcParams['figure.figsize']=(15,10)
#   plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHK', s=50)
#   plt.scatter(bhk3.total_sqft,bhk3.price,marker='+',color='green',label='3 BHK', s=50)
#   plt.xlabel("Total Square Feet Area")
#   plt.ylabel("Price of Per Square Feet")
#   plt.title("location")
#   plt.legend()

# scatter_plot(df7,"Hebbal")

# -------------------------------------------------------------------------------------------------------------
def remove_bhk_outliers(df):   # here we are removing those 2 BHK apartments whose price_per_sqft is less than mean price_per_sqft of 1 BHK apartment
  exclude_indices=np.array([])
  for location, location_df in df.groupby("location"):
    bhk_stats={}
    for bhk,bhk_df in location_df.groupby('BHK'):
      bhk_stats[bhk]={
          "mean":np.mean(bhk_df.price_per_sqft),
          "std":np.std(bhk_df.price_per_sqft),
          'count':bhk_df.shape[0]
      }
    for bhk,bhk_df in location_df.groupby('BHK'):
      stats=bhk_stats.get(bhk-1)
      if stats and stats['count']>5:
        exclude_indices=np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
  return df.drop(exclude_indices,axis='index')

df8=remove_bhk_outliers(df7)
df8.shape #(7329, 7)

# scatter_plot(df8,"Hebbal") # here still abnormality are present so that is fine

# -------------------------------------------------------------------------------------------------------------
# import matplotlib
# matplotlib.rcParams['figure.figsize']=(20,10) # here lies the problem
# plt.hist(df8.price_per_sqft,rwidth=0.8)
# plt.xlabel('Price Per Square Feet')
# plt.ylabel('Count') # till here
# -------------------------------------------------------------------------------------------------------------
df8.bath.unique()
df8[df8.bath>10]

# -------------------------------------------------------------------------------------------------------------
# here we are ploting histogram  for the bathroom
# import matplotlib.pyplot as plt
# plt.hist(df8.bath,rwidth=0.8)
# plt.xlabel("Number Of Bathroom")
# plt.ylabel("Count")
# -------------------------------------------------------------------------------------------------------------
df8[df8.bath>df8.BHK+2]
df9=df8[df8.bath<df8.BHK+2]
df9.shape #(7251, 7)

df10=df9.drop(["size","price_per_sqft"],axis="columns") # here we are removing the size column because we are having BHK column and  also removing "price_per_sqft" because it was just used for outliers detection.
df10.head(3)

dum=pd.get_dummies(df10['location']).astype(int) #applied the one hot encoding on location column.
dum
df11=pd.concat([df10,dum.drop("other",axis="columns")],axis="columns")
df11.head(3)
df12=df11.drop('location',axis="columns") #here it became the final dataframe.
df12.head(3)
#df12.shape #(7251, 245)

X=df12.drop("price",axis="columns") #keeping only independent columns on x
y=df12.price  # keeping dependent columns on y

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)

from sklearn.linear_model import  LinearRegression
linear=LinearRegression()
linear.fit(X_train,y_train)
linear.score(X_test,y_test) #84%




from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                "fit_intercept": [True, False],
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)

def predict_price(location,sqft,bath,bhk):
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return linear.predict([x])[0]

result1=predict_price('1st Phase JP Nagar',1000, 2, 2)
print("The Result is :- ",result1)

result2=predict_price('1st Phase JP Nagar',1000, 3, 3)
print("The Result is :- ",result2)

result3=predict_price('Indira Nagar',1000, 2, 2)
print("The Result is :- ",result3)

import pickle    # here we are saving the model into a file and here there is no actual data
with open("House Price Prediction.pickle","wb") as f:
  pickle.dump(linear,f)

import json  # exporting the columns information in the json
columns={
    'data_columns':[col.lower() for col in X.columns]
}
with open ("columns.json","w") as f :
  f.write(json.dumps(columns))
```
